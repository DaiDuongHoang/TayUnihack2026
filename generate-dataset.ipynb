### Generate Dataset for Clothes Classifier

```python
import csv
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path

from random import uniform
from ultralytics import YOLO
```

```python
# Path to your dataset root folder (contains category subfolders)
dataset_root = Path("../Clothes_Dataset").resolve()

if not dataset_root.exists():
    raise FileNotFoundError(f"Dataset folder not found: {dataset_root}")

print(f"Using dataset folder: {dataset_root}")
```

```python
# Scan subfolders and collect image file records
allowed_ext = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}
records = []

for category_dir in sorted(dataset_root.iterdir()):
    if category_dir.is_dir():
        for image_path in sorted(category_dir.iterdir()):
            if image_path.is_file() and image_path.suffix.lower() in allowed_ext:
                records.append(
                    {
                        "category": category_dir.name,
                        "image_path": str(image_path.resolve()),
                        "filename": image_path.name,
                    }
                )

df = pd.DataFrame(records)
print(f"Total images found: {len(df)}")
print(f"Categories found: {df['category'].nunique() if not df.empty else 0}")
df.head()
```

```python
# Export to CSV and read it back as a DataFrame
csv_path = Path("clothes_dataset_index.csv").resolve()
df.to_csv(csv_path, index=False)
print(f"CSV exported to: {csv_path}")
```
